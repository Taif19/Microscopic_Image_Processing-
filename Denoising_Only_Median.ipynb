**Denoising Image (Only Median Filter)**

```python
from skimage import io, color, img_as_float, img_as_ubyte
from scipy.ndimage import median_filter
import matplotlib.pyplot as plt
from google.colab import files

# -------------------------------
# Step 1: Upload noisy image
# -------------------------------
print("Upload your noisy image:")
uploaded = files.upload()
img_path = list(uploaded.keys())[0]

# -------------------------------
# Step 2: Read and preprocess
# -------------------------------
img = io.imread(img_path)

# Convert RGBA â†’ RGB if needed
if img.ndim == 3 and img.shape[-1] == 4:
    img = img[..., :3]

# Convert to grayscale and normalize
gray = img_as_float(color.rgb2gray(img))

# -------------------------------
# Step 3: Apply Median Filtering
# -------------------------------
# Try different kernel sizes: 3, 5, 7 for comparison
kernel_sizes = [3, 5, 7]
median_imgs = [median_filter(gray, size=k) for k in kernel_sizes]

# -------------------------------
# Step 4: Display results
# -------------------------------
plt.figure(figsize=(16, 6))
plt.subplot(1, len(kernel_sizes) + 1, 1)
plt.imshow(gray, cmap='gray')
plt.title("Original")
plt.axis('off')
plt.imshow(img)

for i, (k, img_f) in enumerate(zip(kernel_sizes, median_imgs), start=2):
    plt.subplot(1, len(kernel_sizes) + 1, i)
    plt.imshow(img_f, cmap='gray')
    plt.title(f"Median size={k}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
