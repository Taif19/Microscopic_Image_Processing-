```python
#Import Libraries

import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.filters.rank import entropy
from skimage.morphology import disk
from skimage.filters import threshold_otsu
from google.colab import files

# Upload Image

uploaded = files.upload()
filename = list(uploaded.keys())[0]
print("Uploaded:", filename)

#Load Image

img = io.imread(filename)
print("Original image:", img.shape)

#Safe Grayscale Conversion

if img.ndim == 3 and img.shape[2] == 4:
    img = img[:, :, :3]   # Remove alpha channel
    print("Converted RGBA to RGB:", img.shape)

#If RGB , convert to grayscale

if img.ndim == 3 and img.shape[2] == 3:
    img_gray = color.rgb2gray(img)
else:
    img_gray = img  # already grayscale

print("Final grayscale:", img_gray.shape)

#grayscale image

plt.figure(figsize=(6,6))
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()


#Otsu Thresholding

thresh = threshold_otsu(img_gray)
binary = img_gray <= thresh

plt.figure(figsize=(6,6))
plt.imshow(binary, cmap="gray")
plt.title("Binary Image (Otsu)")
plt.show()

# ================================
# ðŸ“Œ White Pixel Percentage
# ================================
white_percent = (np.sum(binary == 1) * 100) / binary.size
black_percent= 100-white_percent
print("======================================")
print(f"White pixel region percentage = {white_percent:.2f}%  ")
print("======================================")
print(f"black pixel region percentage = {black_percent:.2f}%")
```

```python
#Import Libraries

import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.filters.rank import entropy
from skimage.morphology import disk
from skimage.filters import threshold_otsu
from google.colab import files

# Upload Image

uploaded = files.upload()
filename = list(uploaded.keys())[0]
print("Uploaded:", filename)

#Load Image

img = io.imread(filename)
print("Original image:", img.shape)

#Safe Grayscale Conversion

if img.ndim == 3 and img.shape[2] == 4:
    img = img[:, :, :3]   # Remove alpha channel
    print("Converted RGBA to RGB:", img.shape)

#If RGB , convert to grayscale

if img.ndim == 3 and img.shape[2] == 3:
    img_gray = color.rgb2gray(img)
else:
    img_gray = img  # already grayscale

print("Final grayscale:", img_gray.shape)

#grayscale image

plt.figure(figsize=(6,6))
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()


#Thresholding (otsu)

thresh = threshold_otsu(img_gray)
binary = img_gray <= thresh

plt.figure(figsize=(6,6))
plt.imshow(binary, cmap="gray")
plt.title("Binary Image (Otsu)")
plt.show()

#Pixel Percentage

white_percent = (np.sum(binary == 1) * 100) / binary.size
black_percent= 100-white_percent
print("======================================")
print(f"White pixel region percentage = {white_percent:.2f}%  ")
print("======================================")
print(f"black pixel region percentage = {black_percent:.2f}%")
```
