```python
!pip install scikit-image scipy matplotlib

from skimage import io, color, img_as_float, img_as_ubyte
from scipy import ndimage as nd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

# Upload image
print("ðŸ“¤ Upload your noisy image:")
uploaded = files.upload()
img_path = list(uploaded.keys())[0]

# Read image
img = io.imread(img_path)

# If image has 4 channels (RGBA), convert to RGB
if img.ndim == 3 and img.shape[-1] == 4:
    img = img[..., :3]

# Convert to grayscale and normalize
gray = color.rgb2gray(img)
gray = img_as_float(gray)

# Step 3: Apply Gaussian filters with multiple sigma values
sigmas = [1, 2, 3, 4, 5]
denoised_imgs = [nd.gaussian_filter(gray, sigma=s) for s in sigmas]

# Step 4: Display comparison
plt.figure(figsize=(16, 8))
plt.subplot(1, len(sigmas) + 1, 1)
plt.imshow(gray, cmap='gray')
plt.title("Original (Noisy)")
plt.axis('off')

for i, (s, img_f) in enumerate(zip(sigmas, denoised_imgs), start=2):
    plt.subplot(1, len(sigmas) + 1, i)
    plt.imshow(img_f, cmap='gray')
    plt.title(f"Gaussian Ïƒ={s}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
from skimage import io, color, img_as_float, img_as_ubyte
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt
from google.colab import files

# -------------------------------
# Step 1: Upload Image
# -------------------------------
print("ðŸ“¤ Upload your noisy image:")
uploaded = files.upload()
img_path = list(uploaded.keys())[0]

# -------------------------------
# Step 2: Read and preprocess
# -------------------------------
img = io.imread(img_path)

# Convert RGBA to RGB if needed
if img.ndim == 3 and img.shape[-1] == 4:
    img = img[..., :3]

# Convert to grayscale and normalize
gray = img_as_float(color.rgb2gray(img))

# -------------------------------
# Step 3: Apply Gaussian denoising
# -------------------------------
sigmas = [1, 2, 3, 4, 5]
denoised_imgs = [gaussian_filter(gray, sigma=s) for s in sigmas]

# -------------------------------
# Step 4: Display results
# -------------------------------
plt.figure(figsize=(18, 6))
plt.subplot(1, len(sigmas) + 1, 1)
plt.imshow(gray, cmap='gray')
plt.title("Original")
plt.axis('off')

for i, (s, img_f) in enumerate(zip(sigmas, denoised_imgs), start=2):
    plt.subplot(1, len(sigmas) + 1, i)
    plt.imshow(img_f, cmap='gray')
    plt.title(f"Ïƒ={s}")
    plt.axis('off')

plt.tight_layout()
plt.show()

```
