```python
#importing

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
from skimage.filters import threshold_sauvola, threshold_niblack
from skimage.morphology import skeletonize

# Upload SEM image
uploaded = files.upload()
file_name = next(iter(uploaded))
img = cv2.imread(file_name)

# Convert to grayscale
if len(img.shape) == 3:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
else:
    gray = img.copy()

print("Image shape:", gray.shape)

# Function image + histogram
def show_image_and_hist(image, threshold=None, title="Histogram"):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    axes[0].imshow(image, cmap='gray')
    axes[0].set_title("SEM Image")
    axes[0].axis('off')
    axes[1].hist(image.ravel(), bins=256, range=(0, 255), color='gray')
    axes[1].set_title(title)
    axes[1].set_xlabel("Pixel Intensity")
    axes[1].set_ylabel("Frequency")
    if threshold is not None:
        axes[1].axvline(int(threshold), color='red', linestyle='--', linewidth=2)
    plt.show()

#Otsu Thresholding

otsu_val, otsu_img = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
print("Otsu Threshold:", otsu_val)
show_image_and_hist(gray, threshold=otsu_val, title="Histogram")
plt.figure(figsize=(6,6))
plt.imshow(otsu_img, cmap='gray')
plt.title("otsu Threshold")
plt.axis('off'); plt.show()

```

```python
#importing

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
from skimage.filters import threshold_sauvola, threshold_niblack
from skimage.morphology import skeletonize

#Upload SEM image
uploaded = files.upload()
file_name = next(iter(uploaded))
img = cv2.imread(file_name)

# Convert grayscale
if len(img.shape) == 3:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
else:
    gray = img.copy()

print("Image shape:", gray.shape)

# Function image + histogram
def show_image_and_hist(image, threshold=None, title="Histogram"):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    axes[0].imshow(image, cmap='gray')
    axes[0].set_title("SEM Image")
    axes[0].axis('off')
    axes[1].hist(image.ravel(), bins=256, range=(0, 255), color='gray')
    axes[1].set_title(title)
    axes[1].set_xlabel("Pixel Intensity")
    axes[1].set_ylabel("Frequency")
    if threshold is not None:
        axes[1].axvline(int(threshold), color='red', linestyle='--', linewidth=2)
    plt.show()

sauvola_thresh = threshold_sauvola(gray)
sauvola_img = (gray > sauvola_thresh).astype(np.uint8) * 255
show_image_and_hist(gray, title="Histogram")
plt.figure(figsize=(4,4))
plt.imshow(sauvola_img, cmap='gray')
plt.title("Sauvola Threshold")
plt.axis('off'); plt.show()
```
